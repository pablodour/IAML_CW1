```python
data_path_train = os.path.join(os.getcwd(), 'data', 'faces_train_data.csv')
data_path_test = os.path.join(os.getcwd(), 'data', 'faces_test_data.csv')


faces_train = pd.read_csv(data_path_train, delimiter = ',')
faces_test = pd.read_csv(data_path_test, delimiter = ',')

X_train = faces_train.drop('smiling', axis=1).values
y_train = faces_train['smiling'].values

X_test = faces_test.drop('smiling', axis=1).values
y_test = faces_test['smiling'].values
```

```python
smiling_x_avg = np.zeros(68)
smiling_y_avg = np.zeros(68)
not_smiling_x_avg = np.zeros(68)
not_smiling_y_avg = np.zeros(68)

smiling_coords = X_train[np.where(y_train == 1)]
not_smiling_coords = X_train[np.where(y_train == 0)]

cols = np.shape(X_train)[1]
for i in range(0, cols, 2):
    smiling_x_avg[i // 2] = np.average(smiling_coords[..., i])
    not_smiling_x_avg[i // 2] = np.average(not_smiling_coords[..., i])


for i in range(1, cols, 2):
    smiling_y_avg[(i - 1) // 2] = np.average(smiling_coords[..., i])
    not_smiling_y_avg[(i - 1) // 2] = np.average(not_smiling_coords[..., i])

plt.plot(not_smiling_x_avg, not_smiling_y_avg, '*', color='orange', label = 'Not smiling');
plt.plot(smiling_x_avg, smiling_y_avg, '*', color='blue', label = 'Smiling');
plt.legend()
plt.savefig('face.png' , bbox_inches='tight', dpi=600)
```
